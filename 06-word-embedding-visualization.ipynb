```python
import gensim
import numpy as np
import matplotlib.pyplot as plt
from gensim.models import KeyedVectors

print('Gensim version:', gensim.__version__)
```

```python
from sklearn.decomposition import PCA

dimred = PCA(n_components=2, copy=False, whiten=True)
```

# Term for Plotting

```python
# Kata-kata yang ingin kita plot vektornya
word_list = ['marah', 'murka', 'muak', 'jijik', 'takut', 'cemas', 'senang', 'bahagia', 'kecewa', 'sedih', 'terpesona', 'terkejut']
```

```python
def plot(data, labels, fc='yellow'):
  # plot the dots
  plt.subplots_adjust(bottom = 0.1)
  plt.scatter(data[:, 0], data[:, 1], marker='o')

  # annotate labels
  for label, x, y in zip(labels, data[:, 0], data[:, 1]):
    plt.annotate(label, \
                 xy=(x, y), xytext=(-15, 15), \
                 textcoords='offset points', ha='right', va='bottom', \
                 bbox=dict(boxstyle='round, pad=0.4', fc=fc, alpha=0.5), \
                 arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0.'))
  plt.show()
```

# IDTweet Word2Vec

```python
word2vec_idtweet_path = 'model-word-embedding/checkpoint/idtweet-word2vec/idtweet-word2vec-300.wordvectors'
word2vec_idtweet = KeyedVectors.load(word2vec_idtweet_path,  mmap='r')

word2vec_idtweet_vec = np.array([word2vec_idtweet[word] for word in word_list])
word2vec_idtweet_vec
```

```python
word2vec_idtweet_dimred = dimred.fit_transform(word2vec_idtweet_vec)
word2vec_idtweet_dimred
```

```python
plot(word2vec_idtweet_dimred, word_list)
```

# IDTweet fastText

```python
fasttext_idtweet_path = 'model-word-embedding/checkpoint/idtweet-fasttext/idtweet-fasttext-300.wordvectors'
fasttext_idtweet = KeyedVectors.load(fasttext_idtweet_path,  mmap='r')

fasttext_idtweet_vec = np.array([fasttext_idtweet[word] for word in word_list])
fasttext_idtweet_vec
```

```python
fasttext_idtweet_dimred = dimred.fit_transform(fasttext_idtweet_vec)
fasttext_idtweet_dimred
```

```python
plot(fasttext_idtweet_dimred, word_list)
```
